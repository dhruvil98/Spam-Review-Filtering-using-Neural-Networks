```
import io
df2 = pd.read_csv(io.BytesIO(uploaded['Labelleddataset.csv']))
df2.head()
```

```
#NN noun singular
#NNS noun plural
#NNP proper noun singular
#NNPS proper noun plural
Label          =[]
hotel_name     =[]
polarity       =[]
source         =[]
review         =[]
noun_proportion=[]
def test():
  Sr_no=0
  for i in df2['text']: #for each Review
    text = i
    blob = TextBlob(text)
    tagged_words=blob.tags
    count=0
    #totl=0
    for word,tag in tagged_words:
    #  totl+=1
      if tag in ["NN","NNS","NNP","NNPS"]:
        count+=1
        # print(f'{word} => {tag}')    
    # if(count/totl>=0.4):
    polarity.append(df2['polarity'][Sr_no])
    Label.append(df2['decision'][Sr_no])
    hotel_name.append(df2['hotel'][Sr_no])
    source.append(df2['source'][Sr_no])
    review.append(df2['text'][Sr_no])
    print(count)
    noun_proportion.append(count/len(i.split()))
    print(f'{Sr_no}=>{df2['polarity'][Sr_no]}=>{df2['decision'][Sr_no]}=>{count/len(i.split())}')
    Sr_no+=1
    
  dataset = {'Label':Label,
             'Polarity':polarity,
             'Hotel':hotel_name,
             'Source':source,
             'Review':review,
             'Noun Proportion':noun_proportion
  }
  df = DataFrame(dataset, columns= ['Label', 'Polarity','Hotel','Source','Review','Noun Proportion'])
  df.to_csv(r'Noun_calculated.csv', index = None, header=True)
```

```
test()
```
